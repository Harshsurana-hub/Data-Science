```python
a = input("Enter a string: ")
reversed_a = ""

for i in range(len(a) - 1, -1, -1):
    reversed_a +=  a[i]

if a == reversed_a:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
lst = [1,2,3,'hello','c',[2,4]]

print(lst)

#append() add single element or list
lst.append(3)

print(lst)

#extend() add multiple elements
lst.extend([3,2])

print(lst)

#insert() add values in between
#lst.insert(index,value)
lst.insert(2,"value")
print(lst)

```

```python
lst = ['poo','esr','gw']
lst.append('Yes')
print(lst)

lst.insert(2,'l')
print(lst)

lst.extend(['l','l'])
print(lst)

#remove()
lst.remove('l')
print(lst)

#pop() removes the last element
lst.pop()
print(lst)

#index()
print(lst.index('esr'))

#count()
print(lst.count('l'))

#sort()
lst.sort()
print(lst)

#reverse()
lst.reverse()
print(lst)
```
