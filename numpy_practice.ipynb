```python
# Question 1:
import numpy as np

# Create a 5×5 matrix of random integers (1–90).
mat = np.random.randint(1, 91, size=(5, 5))
print("matrix : ", mat)

# Find mean of values row and column wise
row_wise_mean = np.mean(mat, axis = 1)
column_wise_mean = np.mean(mat, axis = 0)

print("rowwise mean : ", row_wise_mean)
print("column mean : ", column_wise_mean)

# Replace all even numbers with -1
modified_mat = np.where(mat % 2 == 0, -1, mat)
print("modified : ", modified_mat)

# Sort the array row and column wise
sorted_row = np.sort(mat, axis = 1)
sorted_col = np.sort(mat, axis = 0)

print("sorted row : ", sorted_row)
print("sorted col : ", sorted_col)

# Find row and column wise sum saparately
sum_row = np.sum(mat, axis = 1)
sum_col = np.sum(mat, axis = 0)

print("sum row : ", sum_row)
print("sum col : ", sum_col)

# count the number of -ve's in each column
count_col = np.sum(np.where(modified_mat < 0, 1, 0), axis = 0)
print(count_col)
```
