```python
# Example: Study hours and exam marks

# A teacher thinks students who study more than 5 hours a day get better marks.

# Step 1: Form hypotheses

# Null hypothesis (H₀):
# Studying more than 5 hours a day does not improve exam marks.
# → There is no difference in marks.

# Alternative hypothesis (H₁):
# Studying more than 5 hours a day does improve exam marks.
# → There is an improvement.

# Step 2: Collect data

# You take two groups:

# Group A: studies ≤ 5 hours/day
# Group B: studies > 5 hours/day

# You compare their average marks.

# Step 3: Decision
# If Group B scores are much higher, you may reject H₀
# If scores are similar, you fail to reject H₀
# Simple takeaway
# H₀ = “study time doesn’t matter”
# H₁ = “study time matters”

# If you want, I can give you a coin-flip example or a medical experiment example, which are even easier to understand.
```
