```python
import numpy as np

n = int(input("Number of items: "))

prices = []

for i in range(n):
    prices.append(int(input()))

prices = np.array(prices)

print("Prices =", prices)

print("Total Bill =", np.sum(prices))
print("Average =", np.mean(prices))
print("Minimum =", np.min(prices))
print("Maximum =", np.max(prices))
print("Std Dev =", np.std(prices))

print("Sorted  =", np.sort(prices))
print("Unique  =", np.unique(prices))

print("Items > 100   =", prices[np.where(prices > 100)])
```

```python
stock = np.random.randint(1, 101, size=(5, 5))
print(stock)
demand = np.random.randint(1,101,size=(5,5))
print(demand)
stock_sum_colloum = np.sum(stock,axis=0)
print(stock_sum_colloum)
demand_sum_colloum = np.sum(demand,axis=0)
print(demand_sum_colloum)
remaining_stock = stock-demand
print(remaining_stock)

b = np.where(remaining_stock<0,0,remaining_stock)
print(b)

shortage = remaining_stock - b
print(shortage)
Shortage = np.sum(shortage,axis=0)
print(Shortage)

c = np.where(stock<0,-1,stock)
print(c)

pro_zero = np.sum(c<0)
print(pro_zero)
```
